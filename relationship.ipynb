```python
import pandas as pd
df= pd.read_csv('211210-DRM-new.csv')
```

```python
CH4_Conversion = df.loc[:,'CH4 Conversion']
GHSV= df.loc[:,'GHSV']
Space_Time = 1/GHSV
```

```python
import numpy as np
exp_space_time = -np.exp(-Space_Time)
```

```python
type(exp_space_time)
```

```python
exp_space_time = np.array(exp_space_time).reshape(-1,1)
ch4_conv = np.array(CH4_Conversion).reshape(-1,1)
```

```python
rel = np.concat([exp_space_time,ch4_conv],axis=1)
```

```python
import pandas as pd

df = pd.DataFrame(rel, columns=["min_exp_tau", "X_CH4"])
df_sorted = df.sort_values(by="min_exp_tau")
```

```python
df_sorted
```

```python
import matplotlib.pyplot as plt
plt.plot(df_sorted.iloc[:,0],df_sorted.iloc[:,1])
plt.show()
```
